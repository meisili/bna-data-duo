```python
import pandas as pd
```

```python
# download data and put in the data/ folder. Then change below to data/cdph_data.csv
source_data = "data/CDPH Search results - 1_12_2026, 3_34_15 PM.csv"
cdph = pd.read_csv(source_data, low_memory = False)
# ORIGINAL cdph = pd.read_csv("cdph_data.csv", low_memory = False)
cdph.head()
```

```python
cdph.info()
```

```python
cdph["discontinued_year"] = pd.to_datetime(cdph["Product Discontinued Date"]).dt.year
cdph["product_submitted_year"] = pd.to_datetime(cdph["Product Submitted Date"]).dt.year
cdph["ingredient_submitted_date"] = pd.to_datetime(cdph["Ingredient Submitted Date"]).dt.year
cdph["ingredient_removed_year"] = pd.to_datetime(cdph["Ingredient Removed Date"]).dt.year
```

```python
cdph["discontinued_year"] = cdph["discontinued_year"].astype("Int64")
cdph["product_submitted_year"] = cdph["product_submitted_year"].astype("Int64")
cdph["ingredient_submitted_date"] = cdph["ingredient_submitted_date"].astype("Int64")
cdph["ingredient_removed_year"] = cdph["ingredient_removed_year"].astype("Int64")
```

```python
cdph.head()
```

```python
submissions = cdph.groupby("product_submitted_year").size().rename("product_count").reset_index()
submissions
```

```python
discontinued = cdph.groupby("discontinued_year").size().rename("product_count").reset_index()
discontinued
```

```python
total_products = discontinued["product_count"].sum()
print(total_products)
```

```python
import altair as alt
```

# Discontinued Products by Year

```python
alt.Chart(discontinued).mark_bar().encode(
    x="discontinued_year",
    y="product_count"
)
```

# Submitted Products by Year

```python
alt.Chart(submissions).mark_bar().encode(
    x="product_submitted_year",
    y="product_count"
)
```
