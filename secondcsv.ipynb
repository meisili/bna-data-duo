```python
cdph["discontinued_year"] = pd.to_datetime(cdph["Product Discontinued Date"]).dt.year
cdph["product_submitted_year"] = pd.to_datetime(cdph["Product Submitted Date"]).dt.year
cdph["ingredient_submitted_date"] = pd.to_datetime(cdph["Ingredient Submitted Date"]).dt.year
cdph["ingredient_removed_year"] = pd.to_datetime(cdph["Ingredient Removed Date"]).dt.year
```

```python
cdph["discontinued_year"] = cdph["discontinued_year"].astype("Int64")
cdph["product_submitted_year"] = cdph["product_submitted_year"].astype("Int64")
cdph["ingredient_submitted_date"] = cdph["ingredient_submitted_date"].astype("Int64")
cdph["ingredient_removed_year"] = cdph["ingredient_removed_year"].astype("Int64")
```

# Discontinued Products by Year

```python
alt.Chart(discontinued).mark_bar().encode(
    x="discontinued_year",
    y="product_count"
)
```

```python
pd.set_option('display.max_rows', None)
brands = cdph.groupby("Brand").size().rename("total_count").reset_index()
top_50_brands = brands.sort_values("total_count", ascending = False).head(50)
top_50_brands
```
