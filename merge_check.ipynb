```python
harm_updates = {
    'Cedrus Atlantica Oil/Extract (Cedrus atlantica\xa0oil and extract; Cedrus Atlantica Bark Extract; Cedrus Atlantica Bark Oil; Cedrus Atlantica Bark Water; Cedrus Atlantica Leaf Extract; Cedrus Atlantica Wood Extract; Cedrus Atlantica Wood Oil) 92201-55-3 / 8023-85-6': 'Fragrance Allergen',
    'Alcoholic beverages, when associated with alcohol abuse': 'Hazard Trait Undefined',
    'Myristamide DEA 7545-23-5 ': 'Hazard Trait Undefined',
    'Turpentine gum (Pinus\xa0spp.); Turpentine oil and rectified oil; Turpentine, steam distilled (Pinus\xa0spp.) (Turpentine) 9005-90-7 / 8006-64-2 / 8052-14-0': 'Hazard Trait Undefined',
}

for ingredient, hazard in harm_updates.items():
    final_merge.loc[final_merge['Ingredient Name'].str.contains(ingredient, na=False, regex = False), 'Hazard Traits'] = hazard
```

```python
final_merge.info()
```

```python
final_merge.head()
```

```python
final_merge.to_csv("final_hazard_merge_2_19.csv")
```

```python
hazard_traits_clean = final_merge['Hazard Traits'].str.strip().str.lower().str.replace(r'\s*,\s*', '; ', regex=True)

hazard_dummies = (
    hazard_traits_clean
    .str.get_dummies(sep=';')
    .rename(columns=lambda x: x.strip().title())
)

# Collapse any duplicate columns
hazard_dummies = hazard_dummies.T.groupby(level=0).max().T.astype(bool)
pd.set_option("display.max_columns", None)
final_merge_expanded = pd.concat([final_merge, hazard_dummies], axis=1)
final_merge_expanded
```
