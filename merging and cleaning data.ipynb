```python
# Strip whitespace, split on semicolon, explode into rows
final_merge_explode = final_merge.copy()
final_merge_explode["Hazard Traits"] = final_merge_explode["Hazard Traits"].str.split(";")
final_merge_explode = final_merge_explode.explode("Hazard Traits")
final_merge_explode["Hazard Traits"] = final_merge_explode["Hazard Traits"].str.strip()
final_merge_explode = final_merge_explode.reset_index(drop=True)
```
